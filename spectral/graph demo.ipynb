```python
%pylab inline
import networkx as nx
```

```python
G = nx.Graph()
G.add_node(1)
G.add_nodes_from([2,3])
nx.draw(G)
```

```python
G.add_edge(1, 2, weight=4.7 )
nx.draw(G)
```

```python
nx.draw_random(G)
```

```python
nx.draw_circular(G)
```

```python
nx.draw_spectral(G)
```

```python
H = nx.empty_graph(10)
H.add_edges_from([(1,2),(1,3),(4,5)])
nx.draw(H)
```

```python
H = nx.empty_graph(10)
H.add_weighted_edges_from([(1,2,1),(1,3,2.1),(4,5,0.5)])
nx.draw(H)
```
