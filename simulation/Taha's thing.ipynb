```python
%%R
t<-10000
result<-0
for (i in 1:t){
  scss<-1
  x<-sample(0:6,1)
  ln<-0
  while (ln<6)
  {
    x<-c(x,sample(0:6,1,repl=T))
    scss<-scss+1
    ln<-sum(unique(x)>0)
  }
  result[i]<-scss
}
prob<-0
i<-min(result)
for (i in min(result):max(result))
  {
  prob[i-min(result)+1]<-mean(result==i)*100
}

step<-(1+max(result)-min(result))/8
cut<-seq(min(result),max(result)+1,step)-.5
hist(result,freq = F,breaks=cut,col=rgb(.4,.3,.3))
plot(prob,type = "b",pch=19,col="navy")
plot(ecdf(result),pch=18,col="green")
```
