````python
%%R
set.seed(12)
m <- 100000
aces <- numeric(m)
for (i in 1:m){
    h <- sample(1:52, 5)
    aces[i] <- sum(h<5) 
}

cut = (0:5)-0.5
hist(aces, breaks = cut, prob=T, col="Wheat", main = "Aces in Poker Hands")
cat(summary(as.factor(aces)),"\n")
cat(summary(as.factor(aces))/m, "\n")
cat(round(choose(4,1)*choose(48,4)/choose(52,5),4),"\n")
````
