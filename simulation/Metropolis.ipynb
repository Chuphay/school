```python
m = 500
x = [None for i in range(m)]
y = [None for i in range(m)]
x[0] = 1
y[0] = 3
for i in range(1,m):
    x[i] = x[i-1] + 2*rand()- 1
    y[i] = y[i-1] + 2*rand()- 1
    
figsize(12,5)    
subplot(1,3,1)
plot(x,y); 
subplot(1,3,2)
hist(x);
subplot(1,3,3)
hist(y);
```

```python
m = 500
x = [None for i in range(m)]
y = [None for i in range(m)]
x[0] = 1
y[0] = 3

def accept(newX, newY, oldX, oldY):
    oldF = exp(-(oldX**2 + oldY**2)/2)
    newF = exp(-(newX**2 + newY**2)/2)
    r = newF/oldF
    if(r>1 or r>10*rand()):
        return True
    else:
        return False

acception = []
for i in range(1,m):
    while True:
        tempX = x[i-1] + 2*rand()- 1
        tempY = y[i-1] + 2*rand()- 1
        if(accept(tempX, tempY, x[i-1], y[i-1])):
            acception.append(1)
            x[i] = tempX
            y[i] = tempY
            break
        acception.append(0)    
            
print(sum(acception)/len(acception), var(x), var(y))
figsize(12,5)    
subplot(1,3,1)
plot(x,y); 
subplot(1,3,2)
hist(x);
subplot(1,3,3)
hist(y);
```

Wikipedia has a pretty good article on the Ising Model and MCMC: http://en.wikipedia.org/wiki/Ising_model

```python
from numpy.random import choice
#we ned set up a torus
class Ising(object):
    """Lets do one dimensional"""
    def __init__(self,n, t):
        self.n = n
        self.points = choice([-1,1],n)
        self.temperature = t
    def get_hood(self,i):
        if(i >= self.n): print("fail")
        else:
            return (self.points[(i-1)%self.n], self.points[(i+1)%self.n])
    def get_energy(self):
        H = 0
        for i in range(self.n):
            H +=  self.points[i]*self.points[(i+1)%self.n]  
        return(exp(H/self.temperature))    
    def flip_site(self, i):
        self.points[i] = -self.points[i]
    def get_magnetism(self):
        return abs(sum(self.points))
        
seed(123)
x = Ising(10,1)
print(x.points)
print(x.get_energy())
x.flip_site(1)
x.flip_site(7)
x.flip_site(8)
print(x.points)
print(x.get_energy())
x.points = - x.points
print(x.points)
print(x.get_energy())
```

```python
def accept(newX, oldX):
    r = oldX/newX
    if(r>1 or r>1*rand()):
        return True
    else:
        return False
    
#seed(123)
def MCMC(t,n = 10,m = 100):
    x = Ising(n,t)
    for i in range(1,m):
        while True:
            old_energy = x.get_energy()
            site = randint(0,n)
            x.flip_site(site)
            new_energy = x.get_energy()
            if(accept(old_energy, new_energy)):
                break
            else: 
                x.flip_site(site)
    return(x.get_energy(),x.get_magnetism())           
            
magnetism = []
for i in range(10):
    magnetism.append(MCMC((i+5)/10))
    
plot(magnetism);    
```
