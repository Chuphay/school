```python
%%R
m <- 200
myRandom <- numeric(m)

seed <- 21
myRandom[1] <- seed
  
modulus <- 53; multiplier <- 20; shift <- 0;

for (i in 2:m){ 
    myRandom[i] = (multiplier*myRandom[i-1] + shift)%%modulus
}

myRandom
```

```python
%R plot(myRandom)
```

```python
%%R
rand_generator <-function(seed, num = 53, modulus = 53, multiplier = 20, shift = 0){
    myRandom <- numeric(num)

    myRandom[1] <- seed

    for (i in 2:num){ 
        myRandom[i] = (multiplier*myRandom[i-1] + shift)%%modulus
    }

    myRandom
}
rand_generator(1)
```

```python
%R plot(rand_generator(20, mult = 3))
```

```python
%%R
m <- 100
x <- numeric(m)
y <- numeric(m)
x[1] <- 10
y[1] <- 31
for(i in 2:m){
    x[i] <- rand_generator(x[i-1], mult = 21, num = 2)[2]
    y[i] <- rand_generator(y[i-1], mult = 21, num = 2)[2]
}
plot(y,x)
```

```python
%%R
set.seed(123)
m <- 10000
u <- runif(m)

hist(u, prob = T, ylim= c(0,10))
```

```python
%%R
set.seed(123)
m <- 10000
u <- runif(m)^2
x <- seq(0,1,by=0.001)
hist(u, prob = T, ylim= c(0,10))
lines(x,dbeta(x,0.5,1), col = 'blue', lw = 2)
```
