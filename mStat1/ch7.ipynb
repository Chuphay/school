```python
%%R
hospitals <- read.table("Chapter 7/hospitals.txt", sep =",", header = T)
```

```python
%%R
hist(hospitals$d, breaks = seq(0,3000, by = 200))
```

```python
%%R
mean(hospitals$d)
```

```python
%%R
simulations <- 500
means <- numeric(simulations)
for (i in 1:simulations){
    means[i] <- mean(sample(hospitals$d,8))
}
hist(means,  breaks = seq(0,3000, by = 200))
print(mean(means)); sd(means)
```
