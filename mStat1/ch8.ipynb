```python
from scipy.stats import beta
```

```python
x = linspace(0,1,200)
heads = 13
tot = 20
a = heads + 1
b = tot - heads +1
plot(x, beta.pdf(x,a,b));
```

```python
def plot_this(tot, heads):
    a = heads + 1
    b = tot - heads +1
    plot(x, beta.pdf(x,a,b), label = (tot,heads));
    legend()
plot_this(20,13)    
```

```python
figsize(20,40)
heads = 0
for i in range(1,21):
    if(rand()>0.5):
        heads += 1
    subplot(10,2,i)
    plot_this(i,heads)
```
