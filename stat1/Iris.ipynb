So, to use the $\chi^2$ test (Is this pearson's or something else?) we need to first calculated the expected number of outcomes. 

We will have 6 categories: less than 0.15, 0.15 to 0.25, 0.25 to 0.35, 0.35 to 0.45, 0.45 to 0.55, and more than 0.55.

Using $\mu$ and $\sigma$ from above (Is that best? there may be better $\mu$ and $\sigma$ that fits the data better, but let's assume its ok) we have $\mu = 0.24400$ and $\sigma = 0.10721$

So, we have to figure out the probabilities, for example, the first one is like $\mathbb{P}(X \le 0.15)$ and the next one will be like $\mathbb{P}(0.15 \le X \le 0.25)$

```python
mu = 0.244
std = 0.10721
v = [pnorm(0.15,mu,std), pnorm(0.25,mu,std) - pnorm(0.15,mu,std),
      pnorm(0.35,mu,std) - pnorm(0.25,mu,std), pnorm(0.45,mu,std) - pnorm(0.35,mu,std),
      pnorm(0.55,mu,std) - pnorm(0.45,mu,std), 1 - pnorm(0.55,mu,std)]
v
```

Then we have to multiply these values by 50 to get the expected number. From there it's fairly simple to calculate a $\chi^2$ test. 

```python
count = [0,0,0,0,0,0]
for i in df[:50][3]:
    if(i<0.15): count[0] += 1
    if(0.15<i<0.25): count[1] += 1
    if(0.25<i<0.35): count[2] += 1
    if(0.35<i<0.45): count[3] += 1
    if(0.45<i<0.55): count[4] += 1
    if(0.55<i): count[5] += 1    
        
count        
```

```python
mu = 0.22
std = 0.092
v = [pnorm(0.15,mu,std), pnorm(0.25,mu,std) - pnorm(0.15,mu,std),
      pnorm(0.35,mu,std) - pnorm(0.25,mu,std), pnorm(0.45,mu,std) - pnorm(0.35,mu,std),
      pnorm(0.55,mu,std) - pnorm(0.45,mu,std), 1 - pnorm(0.55,mu,std)]
list(map(round, (50*array(v))))
```

```python
chi_sq_int = (array(count)-50*array(v))**2/(50*array(v))
print(chi_sq_int)
print(sum(chi_sq_int))
```
