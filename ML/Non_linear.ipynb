```python
def plot_data(data, target):
    for i, e in enumerate(data):
        #print(i)
        if(target[i] == 1):
            plot(data[i][0],data[i][1],c = 'r', marker = 'x')
        else:
            plot(data[i][0],data[i][1],c = 'b', marker = 'o')
def plot_perceptron(w, c = 'k'):
    x = linspace(0,1,100)
    slope = -w[1]/w[2]
    intercept = - w[0]/w[2]
    plot(x,slope*x +intercept,c)
```

```python

seed(131)
length = 40
data = []
target = [-1 for i in range(length)]
for i in range(length):
    x = 2*rand(2) - 1
    data.append(x)
    if(x[0]**2 + x[1]**2<0.4):
        target[i] = 1
    #if(rand()<0.01): #add some noise
    #    target[i] = randint(0,2)
        
for i in range(length):
    if target[i] == 1:
        scatter(data[i][0],data[i][1],c = 'r')
    else:
        scatter(data[i][0],data[i][1], c= 'b')
```

```python
new_data = []
for i, e in enumerate(data):
    x,y = e
    new_data.append([x**2,y**2])

for i in range(length):
    if target[i] == 1:
        scatter(new_data[i][0],new_data[i][1],c = 'r')
    else:
        scatter(new_data[i][0],new_data[i][1], c= 'b')   
        
        
```

```python

new_data = array(new_data)
def linear_regression(data, target):
    
    x = hstack((np.ones((data.shape[0],1)),data))
    w = dot(pinv(x), target)
    return w


def PLA(data,target,max = 200):
    w = randn(3)
    
    for i in range(max): #just in case the data is not linearly seperable
        error = False
        for i, e in enumerate(data):
            x = array([1,e[0],e[1]])

            if(sign(dot(w,x)) != target[i]):          
                error = True
                w = w + target[i]*x
       
        if(error == False):
            print("success!")
            break
    return w        
pla = PLA(new_data,target) 
w =  linear_regression(new_data, target)
plot_data(new_data,target)
plot_perceptron(pla,'r')    
plot_perceptron(w) 
```

```python
seed(123)
length = 40
data = [rand(2) for i in range(length)]
target = []
for i in range(length):
    if(data[i][0] + data[i][1]< 1):
        target.append(1)
    else:
        target.append(-1)
        
plot_data(data,target)    
w =  linear_regression(array(data), target)
w_pla  = PLA(array(data),target)
plot_perceptron(w_pla,'b')
plot_perceptron(w,'r')  
```
