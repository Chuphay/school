```python
data_5 = []
data_1 = []
for pic in data[:-1]: #avoiding errors?
    image = array(list(map(float,pic.split())))
    target = image[256:]
    image = image[:256]
    image = image.reshape(16,16)
    if(target[5]):
        data_5.append(image)
    if(target[1]):
        data_1.append(image)
```

```python
figsize(3,3)
image = data_5[16]
imshow(image, interpolation='nearest').set_cmap('gray')
```

```python
def get_intensity(image):
    out = 0.0
    for i in image:
        for j in i:
            if j == 1:
                out +=1
    return out/256  

def get_symmetry(image):
    out = 0.0
    for i in range(8):
        for j in range(8):
            if(image[i][j] == image[15-i][j]):
                out += 1               
    return out/256
```

```python
data_summary = []
for i in data_1:
    data_summary.append([get_intensity(i), get_symmetry(i),1])
for i in data_5:
    data_summary.append([get_intensity(i), get_symmetry(i),-1])
data_summary[0]   
```

```python
figsize(8,6)
for i, e in enumerate(data_summary):
    #print(i)
    if(data_summary[i][2] == 1):
        plot(data_summary[i][0],data_summary[i][1],c = 'r', marker = 'x')
    else:
        plot(data_summary[i][0],data_summary[i][1],c = 'b', marker = 'o')
```
