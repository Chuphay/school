```python
points = 20
data = {}
intercept = -3
slope = 5
for i in range(points):
    x = 5*rand()
    y = 5*rand()
    y_prime = slope*x + intercept
    if(y_prime>y):
        data[i] = [x,y,1]
    else:
        data[i] = [x,y,-1]
```

```python
for i in data:
    if(data[i][2] == -1):
        scatter(data[i][0],data[i][1], c = 'r', marker = 'x')
    else:
        scatter(data[i][0],data[i][1], c = 'b', marker = 'o')        
```

```python
w = randn(3)
for i in data:
    if(data[i][2] == -1):
        scatter(data[i][0],data[i][1], c = 'r', marker = 'x')
    else:
        scatter(data[i][0],data[i][1], c = 'b', marker = 'o')
def plot_perceptron(w):
    x = linspace(0,5,100)
    slope = - w[1]/w[2]
    intercept = - w[0]/w[2]
    plot(x,slope*x +intercept)
plot_perceptron(w)    
```

```python
def classify(w,x):
    x = insert(x,0,1)
    if(dot(w,x)>0):
        return 1
    else:
        return -1   
```

```python
MAX = 100
for i in range(MAX):
    x = int(points*rand())
    if(classify(w,data[x][:2]) == data[x][2]):
        pass
    else:      
        w = w + data[x][2]*array([1,data[x][0],data[x][1]])
        for i in data:
            if(data[i][2] == -1):
                scatter(data[i][0],data[i][1], c = 'r', marker = 'x')
            else:
                scatter(data[i][0],data[i][1], c = 'b', marker = 'o')
        plot_perceptron(w)         
    
```
