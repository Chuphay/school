```python
def get_two_points():
    x = 2*rand(2) - 1
    y = sin(pi*x)
    return array([x,y])

def get_line_parameters(data, plot_this = False):
    slope = (data[1][0] - data[1][1])/(data[0][0] - data[0][1])
    intercept = data[1][1] - slope*data[0][1]
    if(plot_this):
        x = linspace(-1,1,100)
        plot(x, slope*x+intercept)
    return (slope, intercept)    
  
for i in range(100):
    data = get_two_points()    
    scatter(data[0], data[1]);
    get_line_parameters(data, True)
```

```python
x = linspace(-1,1,100)
trapz(sin(pi*x)**2,x)
```

```python
def get_average(data, plot_this = False):
    intercept = (data[1][0] + data[1][1])/2
    if(plot_this):
        x = linspace(-1,1,100)
        y = 0*x + intercept
        plot(x, y)
    return intercept

num = 20000
m = 0
var = 0
for i in range(num):
    data = get_two_points()    
    #scatter(data[0], data[1]);
    avg =  get_average(data, False)
    var += avg**2
    m += avg
avg_intercept = m/num
print("var",var/num)
print("bias",trapz((sin(pi*x) - avg_intercept)**2,x)/2)    
```
