Let X and Y have a bivariate normal distribution with $\mu_X = 70$, $\sigma_X^2 = 100$, $\mu_Y = 80$, $\sigma_Y^2 = 169$, and $\rho = 5/13$. Find

(a) E(Y | X = 72).

The book supplies us with the following formula

$$\mathbb{E}(Y|x) = \mu_Y + \rho \frac {\sigma_Y}{\sigma_X} (x - \mu_X)$$

Thus, 

$$\mathbb{E}(Y|X = 72) = 80 + (5/13)*(13/10)*(72-70) = 81$$  

(b) Var(Y | X = 72).

Again, the book supplies us with the following formula

$$\sigma^2_{Y|x} = \sigma^2_Y(1-\rho^2) = 169*(1 - 25/169) = 144$$

(c) P(Y ≤ 84, | X = 72).

Here, we use the fact that given that X = 72, the mean of Y is 81, with standard deviation 12, Thus $Y \sim N(81, 144) $

$$\mathbb{P}(Y \leq 84 | X = 72) = \mathbb{P}(\frac {Y-81} {12} \leq \frac {84-81} {12} | X = 72) = 0.5987063$$

Let

$$f(x,y) = \left(\frac 1 {2 \pi} \right)e^{ −(x^2 +y^2 )/2}\left[ 1 + xye^{ −(x^2 +y^2 −2)/2} \right]$$

$$−\infty < x <\infty, −\infty < y < \infty.$$

Show that f (x, y) is a joint pdf and the two marginal pdfs
are each normal. Note that X and Y can each be normal,
but their joint pdf is not bivariate normal.

To show that it is a joint pdf we will first show that the integration over the domain is equal to one. We first split the integrand into two parts:

$$f(x,y) = \left(\frac 1 {2 \pi} \right)e^{ −(x^2 +y^2 )/2} + \left(\frac 1 {2 \pi} \right)xye^{ −(x^2 +y^2 −1)}$$

The second integrand is odd and therefore zero over the domain. The first integrand is one by recalling the fact that

$$\int \frac 1 {\sqrt{2\pi}} e^{ −x^2/2} = 1$$ 

Next, we must show that the probability is positive. To do this, it is enough to look at the minimum value of this function:

$$1+ xe^{-x^2/2}$$

Taking the derivative and setting it to zero, yields:

$$0 = e^{-x^2/2} - x^2e^{-x^2/2}$$

which gives us a minimum at negative one. Plugging that in to the equation above yields a positive quantity. Showing that the probability is always less than one should be obvious both from inspection and because of the fact that the integral over the whole domain is equal to one.

Finding the marginal pdfs is straightforward. Using the facts above, it is obvious that

$$f_1(x) = \frac 1 {\sqrt{2\pi}} e^{ −x^2/2} $$

$$f_2(y) = \frac 1 {\sqrt{2\pi}} e^{ −y^2/2} $$

as required.

$$f (x) = \frac {e^{−x}}{(1 + e^{ −x} )^2} \hspace{15 mm} -\infty<x <\infty$$

$$Y = \frac 1 {1 + e^{−X}}$$

has a U(0, 1) distribution.

$u(x) = \frac 1 {1 + e^{−x}}$ and  $\nu(y) = \ln \left(\frac y {1-y} \right)$.
