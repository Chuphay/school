<h3>3.5</h3>  
<b>c.)</b>

```python
%%R
x <- c(1,0,2,0,3,1,0,1,2,0)
y <- c(16, 9, 17, 12, 22, 13, 8, 15, 19, 11)
b_1 <- cov(x,y)/var(x)
b_0 <- mean(y) -b_1*mean(x)
resids <- y - b_0 -b_1*x
stem(resids)
cat('A Stem-and-leaf plot is essentially a histogram, but the numbers of the data are also displayed.\n')
```

<b>d.)</b>

```python
%R plot(x,resids)
```

It appears to me that the residuals are evenly distributed around zero for all X.

<b>e.)</b>

```python
%%R
qqnorm(resids)
cat(sort(resids),'\n')
MSE <- sum(resids^2)/8
at_where <- (1:10-0.375)/(10+0.25) #see (3.6)
E <-sqrt(MSE)*qnorm(at_where) 
cat(E,'\n\n')

Sxy = sum((E-mean(E))*(sort(resids) - mean(resids)))
Sxx = sum((E-mean(E))^2)
Syy = sum((resids - mean(resids))^2)
cat('correlation r:',Sxy/sqrt(Sxx*Syy), cor(sort(resids),E),'\n')
cat('The appendix has the value 0.879, since r is above this we accept the null hypothesis')
```

<b>f.)</b>

```python
%R plot(resids, type = 'l')
```

Because there is no systematic pattern, the data is random with respect to time.

<b>g.)</b>

```python
%%R
fit <- lm(resids^2~x)
coeff <-coef(fit)
b_0_res = coeff[1]
b_1_res = coeff[2]
SSR = b_1_res^2*sum((x-mean(x))^2)
cat('SSR: ',SSR,'\n')
SSE = sum((resids^2 - b_0_res - b_1_res*x)^2)
cat('SSE: ', SSE,'\n')
cat('Breusch-Pagen number is',(SSR/2)/(SSE/10)^2)
cat(' which is bigger than ',qchisq(.9, df=1),'. We therefore conclude the variance is constant.\n') 
```
