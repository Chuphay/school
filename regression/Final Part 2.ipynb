```python
%%R
data <- read.table("categorical.csv", sep = ",", header = T)
```

```python
%%R
samples <- sample(length(data[,1]),50)
data_test <- data[samples,]
data_train <- data[-samples,]
fit <- lm(lwage ~ educ+educsqr+exper+expersq+log.tenure+
          as.factor(numdep)+as.factor(race)+as.factor(metro)+
          as.factor(marital.st)+as.factor(gender)+
          as.factor(business.type)+as.factor(job.type)+
          educ:exper+ as.factor(metro):log.tenure+ as.factor(gender):as.factor(race)+
          as.factor(gender):as.factor(marital.st)+ as.factor(gender):as.factor(business.type)+
          as.factor(gender):as.factor(job.type)+as.factor(gender):as.factor(metro), data = data_train)
library("MASS")
stepAIC(fit)$anova
```

```python
%%R
fit_train <- lm(lwage ~ educ + educsqr + exper + expersq + log.tenure + as.factor(metro) + 
    as.factor(marital.st) + as.factor(gender) + as.factor(business.type) + 
    as.factor(job.type) + log.tenure:as.factor(metro) + as.factor(marital.st):as.factor(gender), data = data)
summary(fit_train)
```

```python
%%R
wage_prediction <- data.frame( data_test$wage, exp(predict(fit_train, data_test)), 
                              abs(data_test$wage - exp(predict(fit_train, data_test))),
                             data_test$lwage , predict(fit_train, data_test),
                             abs(data_test$lwage - (predict(fit_train, data_test))))
colnames(wage_prediction) <- c('real','pred','diff', 'lwage', 'lpred','ldiff')
for (i in seq(0.2,3.6,by=0.01)){
    cat(i,sum(wage_prediction$diff < i)/length(wage_prediction$diff),'\n')
        #sum(wage_prediction$diff < exp(i))/length(wage_prediction$diff),'\n')
    }
```
